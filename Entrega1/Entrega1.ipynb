Lea la imagen "col.png" y la imagen "jap.png" que se encuentran en la carpeta "res", muestrelas con plt.imshow.

```python
#se leen las dos imagenes
#se transforman del espacio BGR a RGB
col = cv2.imread('res/col.png',cv2.IMREAD_COLOR)
col = cv2.cvtColor(col,cv2.COLOR_BGR2RGB)
plt.imshow(col);
```

```python
jap = cv2.imread('res/jap.png',cv2.IMREAD_COLOR)
jap = cv2.cvtColor(jap,cv2.COLOR_BGR2RGB)
plt.imshow(jap);
```

<br> A continuacin, se aplica la transformacin sobre ambas imgenes. </br>

# <center> $ G(x,y) = (1-k)*I1(x,y) + k*I2(x,y) $ </center>

# <center> $ 0 < k < 1 $ </center>

```python
#se debe reescalar la imagen porque tienen tamanos diferentes
col = cv2.resize(col,(250,166))
```

```python
#funcion que define la transformacion
def transform1(i1,i2,k):
    
    a = np.multiply(i1,1-k)
    b = np.multiply(i2,k)
    
    result = np.add(a,b)
    result = result.astype(np.uint8)
    return result
```

```python
#se realiza la transformacion

#parametro de la funcion
# k [0,1]
k = 0.5

transformada = transform1(col,jap,k)
fig, (ax1,ax2,ax3) = plt.subplots(1,3,figsize=(15,3))
ax1.set_title('Imagen 1')
ax2.set_title('Imagen 2')
ax3.set_title('Transformacion')
ax1.imshow(col)
ax2.imshow(jap)
ax3.imshow(transformada);
```

<p>Que ocurre cuando k vale 0, 0.5 y 1.</p>
<p>La transformacin bsicamente combina dos imgenes tomando valores por cada pixel y asigna finalmente un valor proporcional a la constante k. Un valor de 0 en k hace que la imgen 1 sea la salida de la funcin y un valor de 1 har lo mismo para la imagen 2. As que un valor de 0.5 tomar los valores de ambas imgenes y devolver el promedio. Efectivamente generando un efecto de "fusin" de las imgenes.</p>

<h4><span style="color: #000000;">1. b)&nbsp;</span></h4>
<p>Se aplica la siguiente transformacin</p>

# <center> $ G(x,y) = a*I1(x,y) + b $ </center>

```python
#se define la transformacion a*i + b
def transform2(i,a,b):
    s = np.multiply(i,a)
    result = np.add(s,b)
    result = result.astype(np.uint8)
    return result
```

```python
#parametros de la funcion
a = 2
b = 0
transformada = transform2(col,a,b)

fig, (ax1,ax3) = plt.subplots(1,2,figsize=(10,3))
ax1.set_title('Imagen')
ax3.set_title('Transformacion')
ax1.imshow(col)
ax3.imshow(transformada);
```

<p>El valor a de la funcin multiplica cada pixel por ese nmero y luego se le suma el valor b. En particular para el caso donde a es 0 y b es cualquier valor, a cada pixel se le asigna el valor b despus de la transformacin. El valor b determinar si la imgen es negra, blanca o una gama intermedia de grises si a es 0.</p>

```python
%matplotlib inline

import cv2
from matplotlib import pyplot as plt
import numpy as np

#Leer la imagen de entrada
img1 =cv2.cvtColor(cv2.imread('res/gato.jpg'), cv2.COLOR_BGR2RGB)

#Utilizamos la funcin cv2.Canny, que corresponde a un algoritmo de deteccin de bordes
edges = cv2.Canny(img1,100,200)
#Resaltar los bordes de blanco si sus pixeles son menores que un valor umbral
edges[edges<128] = 0   
#Convertir la imagen edges a RGB
edges = cv2.cvtColor(edges, cv2.COLOR_GRAY2RGB)

edges[:,:,1]=0               #para obtener el color magenta (255,0,255)

#Crear una figura 1x3 para mostrar la imagen de entrada, la imagen obtenida con edges y la superposicin
fig = plt.figure(figsize=(15,24))

plt.subplot(1,3,1)
plt.imshow(img1,cmap='gray')
plt.title("Imagen 1")

plt.subplot(1,3,2)
plt.imshow(edges,cmap='gray')
plt.title("Imagen 2")

plt.subplot(1,3,3)
plt.imshow(cv2.add(img1,edges),cmap='gray')
plt.title("Imagen 3");
```
