```python
%matplotlib inline
#importar librerias relevantes
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

<h4><span style="color: #000000;">1. a)&nbsp;</span></h4>
<p>&nbsp;Generar una im&aacute;gen, mostrarla y explicar c&oacute;mo se gener&oacute;.</p>

```python
#imagen en negro 400x400
img = np.zeros((399,399,3),np.uint8)
#franja amarilla
#(R,G,B)=(255,255,0)
img[:,0:100,0] = 255
img[:,0:100,1] = 255
#franja cyan
#(R,G,B)=(0,255,255)
img[:,100:200,1] = 255
img[:,100:200,2] = 255
#franja roja
#(R,G,B)=(255,0,0)
img[:,200:300,0] = 255
#franja verde
#(R,G,B)=(0,255,0)
img[:,300:400,1] = 255

plt.imshow(img);
```

<p>Para generar la im&aacute;gen:</p>
<p>Se genera un arreglo de 400x400x3. 400x400 para la dimensi&oacute;n de la im&aacute;gen en pixeles en 3 canales.</p>
<p>Las franjas de colores son amarilla, cyan, roja y verde. Se pueden generar con diferentes combinaciones de los colores principales (rojo, verde y azul) en m&aacute;ximo nivel de saturaci&oacute;n.</p>

<p>Amarillo: (255,255,0)</p>
<p>Cyan: (0,255,255)</p>
<p>Verde: (0,255,0)</p>
<p>Rojo: (255,0,0)</p>
<p>Se asignan los valores en rangos de 400x100 pixeles.</p>

<h4><span style="color: #000000;">1. b)&nbsp;</span></h4>
<p>Leer la im&aacute;gen en los canales R,G y B del espacio RGB y los canales L,A y B del espacio LAB con sus respectos histogramas.</p>

```python
#separar la imagen en canales RGB
imgR = img[:,:,0]
imgG = img[:,:,1]
imgBlue = img[:,:,2]

#convertir la imagen en LAB
imglab = cv2.cvtColor(img,cv2.COLOR_RGB2LAB)
#separar los canales LAB
imgL = imglab[:,:,0]
imga = imglab[:,:,1]
imgb = imglab[:,:,2]

#canales RGB y LAB separados
fig, (ax1,ax2) = plt.subplots(2,3,figsize=(15,7))
fig.suptitle('RGB y LAB',fontsize=20)
ax1[0].set_title('R')
ax1[0].imshow(imgR,cmap='Reds',aspect='auto')
ax1[1].set_title('G')
ax1[1].imshow(imgG,cmap='Greens',aspect='auto')
ax1[2].set_title('B')
ax1[2].imshow(imgBlue,cmap='Blues',aspect='auto')
ax2[0].set_title('L')
ax2[0].imshow(imgL,cmap='gray',aspect='auto')
ax2[1].set_title('A')
ax2[1].imshow(imga,cmap='gray',aspect='auto')
ax2[2].set_title('B')
ax2[2].imshow(imgb,cmap='gray',aspect='auto');
```

```python
plt.clf()
fig,(ax1,ax2) = plt.subplots(2,3,figsize=(15,7))
fig.suptitle('Histogramas RGB y LAB')
ax1[0].set_title('R')
ax1[0].hist(imgR.ravel(),bins = 255, range = (0.0,255.0))
ax1[1].set_title('G')
ax1[1].hist(imgG.ravel(),bins = 255, range = (0.0,255.0))
ax1[2].set_title('B')
ax1[2].hist(imgBlue.ravel(),bins = 270, range = (0.0,270.0))
ax2[0].set_title('L')
ax2[0].hist(imgL.ravel(),bins = 255, range = (0.0,255.0))
ax2[1].set_title('A')
ax2[1].hist(imga.ravel(),bins = 255, range = (0.0,255.0))
ax2[2].set_title('B')
ax2[2].hist(imgb.ravel(),bins = 255, range = (0.0,255.0));
```

<h3>Conclusiones</h3>
<p>Para esta imágen en particular, la división en RGB habla de una información que conocíamos, pues muestra la composición de cada franja en cada canal del espacio RGB.</p>
<p>El espacio LAB: el canal L indica la luminosidad del color, la predominancia de valores altos indica alta luminosidad. Los canales a y b tienen información del color. Valores altos en el canal b indican predominancia de amarillos sobre azules y valores bajos en a indican predominancia en verdes sobre rojos.</p>

<p>Lea la imagen "eiffel.jpg" que se encuentra en la carpeta "res", redimensionela a 800x600 p&iacute;xeles, muestre la imagen y su histograma asociado.</p>

```python
#Imagen original e histograma de los canales RGB
eiffel = cv2.imread('res/eiffel.jpg',cv2.IMREAD_COLOR)
eiffel = cv2.cvtColor(eiffel,cv2.COLOR_BGR2RGB)
eiffel = cv2.resize(eiffel,(800,600))
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(10,5))
colors = ('r','g','b')
fig.suptitle('Imagen e Histograma')
ax1.set_title('Imagen')
ax1.imshow(eiffel,aspect='auto')
ax2.set_title('Histograma')
for i, col in enumerate(colors):
    imgarray = eiffel[:,:,i].ravel()
    ax2.hist(imgarray,histtype='step',bins=255,
             range=(0.0,255.0),normed=True,color=colors[i])
```

<h3>Conclusiones</h3>
<p>El canal H del HSV brinda información del color derivando del círculo cromático. La alta concentración en el histograma indica una alta presencia de un color o por lo menos gamas de este color. Esto se respalda en el histograma de los canales RGB que muestran alta presencia del color azul. El canal S habla de la saturación. <font size = 6 color = blue><b><center> MÓDULO II </center></b></font>

Lea la imagen "col.png" y la imagen "jap.png" que se encuentran en la carpeta "res", muestrelas con plt.imshow.

```python
#se leen las dos imagenes
#se transforman del espacio BGR a RGB
col = cv2.imread('res/col.png',cv2.IMREAD_COLOR)
col = cv2.cvtColor(col,cv2.COLOR_BGR2RGB)
plt.imshow(col);
```

```python
jap = cv2.imread('res/jap.png',cv2.IMREAD_COLOR)
jap = cv2.cvtColor(jap,cv2.COLOR_BGR2RGB)
plt.imshow(jap);
```

<br> A continuación, se aplica la transformación sobre ambas imágenes. </br>

# <center> $ G(x,y) = (1-k)*I1(x,y) + k*I2(x,y) $ </center>

# <center> $ 0 < k < 1 $ </center>

```python
#se debe reescalar la imagen porque tienen tamanos diferentes
col = cv2.resize(col,(250,166))
```

```python
#funcion que define la transformacion
def transform1(i1,i2,k):
    
    a = np.multiply(i1,1-k)
    b = np.multiply(i2,k)
    
    result = np.add(a,b)
    result = result.astype(np.uint8)
    return result
```

```python
#se realiza la transformacion

#parametro de la funcion
# k [0,1]
k = 0.5

transformada = transform1(col,jap,k)
fig, (ax1,ax2,ax3) = plt.subplots(1,3,figsize=(15,3))
ax1.set_title('Imagen 1')
ax2.set_title('Imagen 2')
ax3.set_title('Transformacion')
ax1.imshow(col)
ax2.imshow(jap)
ax3.imshow(transformada);
```

<p>Que ocurre cuando k vale 0, 0.5 y 1.</p>
<p>La transformación básicamente combina dos imágenes tomando valores por cada pixel y asigna finalmente un valor proporcional a la constante k. Un valor de 0 en k hace que la imágen 1 sea la salida de la función y un valor de 1 hará lo mismo para la imagen 2. Así que un valor de 0.5 tomará los valores de ambas imágenes y devolverá el promedio. Efectivamente generando un efecto de "fusión" de las imágenes.</p>

<h4><span style="color: #000000;">1. b)&nbsp;</span></h4>
<p>Se aplica la siguiente transformación</p>

# <center> $ G(x,y) = a*I1(x,y) + b $ </center>

```python
#se define la transformacion a*i + b
def transform2(i,a,b):
    s = np.multiply(i,a)
    result = np.add(s,b)
    result = result.astype(np.uint8)
    return result
```

```python
#parametros de la funcion
a = 2
b = 0
transformada = transform2(col,a,b)

fig, (ax1,ax3) = plt.subplots(1,2,figsize=(10,3))
ax1.set_title('Imagen')
ax3.set_title('Transformacion')
ax1.imshow(col)
ax3.imshow(transformada);
```

<p>El valor a de la función multiplica cada pixel por ese número y luego se le suma el valor b. En particular para el caso donde a es 0 y b es cualquier valor, a cada pixel se le asigna el valor b después de la transformación. El valor b determinará si la imágen es negra, blanca o una gama intermedia de grises si a es 0.</p>

```python
%matplotlib inline

import cv2
from matplotlib import pyplot as plt
import numpy as np

#Leer la imagen de entrada
img1 =cv2.cvtColor(cv2.imread('res/gato.jpg'), cv2.COLOR_BGR2RGB)

#Utilizamos la función cv2.Canny, que corresponde a un algoritmo de detección de bordes
edges = cv2.Canny(img1,100,200)
#Resaltar los bordes de blanco si sus pixeles son menores que un valor umbral
edges[edges<128] = 0   
#Convertir la imagen edges a RGB
edges = cv2.cvtColor(edges, cv2.COLOR_GRAY2RGB)

edges[:,:,1]=0               #para obtener el color magenta (255,0,255)

#Crear una figura 1x3 para mostrar la imagen de entrada, la imagen obtenida con edges y la superposición
fig = plt.figure(figsize=(15,24))

plt.subplot(1,3,1)
plt.imshow(img1,cmap='gray')
plt.title("Imagen 1")

plt.subplot(1,3,2)
plt.imshow(edges,cmap='gray')
plt.title("Imagen 2")

plt.subplot(1,3,3)
plt.imshow(cv2.add(img1,edges),cmap='gray')
plt.title("Imagen 3");
```

<p> CONLUSIONES </p>
 
<p>Se logra establecer una relacion pixel a pixel, mediante operaciones de suma con imagenes, produciendo una imagen resultante la cual da un efecto visual de superposicion; en este caso el orden de los factores no altera el resutado por lo que si se intercambia la entrada de cada imagen en la transformada esta no se vera alterada ya que es una transfomacion lineal.
    
El parametro a el cual multiplica la matriz por un escalar aumenta o dismiuye el contraste, influyendo en los tres canales que generan los colores de la imagen segun el valor dado, realizando una distribucion amplia del histograma del histograma, el parametro b representa el desplazamiento del histograma a niveles altos o bajos de intensidad generando una imagen muy luminosa o muy opaca.
se pueden obtener contornos de imagen para realizar cambios o mejoras en la imagen resultante, y con las operacciones suma se observa la superposicion de la imagen original y el contorno. </p>
