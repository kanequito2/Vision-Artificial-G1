```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
import cv2
```

```python
img = cv2.imread("owo.bmp", 0)
```

```python
plt.imshow(img)
```

```python
print(img.shape)
print(img.shape[0])
print(img.shape[1])
```

```python
len_x = img.shape[0]
len_y = img.shape[1]

z = img.flatten()
z = z.reshape((len(z),1))
y = np.array([ i for i in range(len_y) ]*len_x)
x = np.array([ [i]*len_y for i in range(len_x)  ])
x = x.flatten()
xx = np.power(x, 2)
yy = np.power(y, 2)
xy = x*y
ones = np.ones(len(z))
H = np.array([ xx, yy, xy, x, y, ones ])
H = H.T
c = np.dot(H.T, H)
c = np.linalg.inv(c) 
c = np.dot(c, H.T)
c = np.dot(c, z)
z_fit = np.dot(H, c)
z_fit = z_fit.reshape((len_x, len_y)).astype(np.uint8)
```

```python
plt.imshow(z_fit, cmap='gray')
print(img.dtype)
print(z_fit.dtype)
```

```python
plt.figure(figsize=(10,10), dpi = 100)
res = cv2.subtract(img, z_fit)
print(np.max(res))
print(np.min(res))
plt.imshow(res * (res < 160), cmap="hot")
```
